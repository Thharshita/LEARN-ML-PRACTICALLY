```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing dataset

```python
dataset = pd.read_csv('../Big Basket.com Cart Apriori.csv', header = None)
transactions = []
print(dataset)

for i in range(0, 7):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])

#For numerical data: You can work directly with the DataFrame, utilizing Pandas' capabilities for numerical data analysis.
#For categorical or binary data: You might consider converting the data into a transaction format if you intend to
#use association rule mining algorithms like Apriori to discover interesting patterns.


#The code transaction.append([str(dataset.values[i, j]) for j in range(0, 20)]) generates a new list for each row because of the list comprehension used inside the append() function.


display(transactions)
```

## Apriori Training on Dataset

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualizing

### Raw Results

```python
results = list(rules)
```

### Proper Display

```python
"""RelationRecord(
    items=frozenset({'burgers', 'almonds'}),
    support=0.005402410306136584,
    ordered_statistics=[
        OrderedStatistic(
            items_base=frozenset({'almonds'}),
            items_add=frozenset({'burgers'}),
            confidence=0.26530612244897955,
            lift=3.0594966421073218
        )
    ]
)
"""


"""result: This refers to a single RelationRecord object, representing the result of one association rule.
result[2]: This accesses the list of ordered statistics within the RelationRecord object. Each ordered statistic represents
an association rule.

result[2][0]: This accesses the first ordered statistic in the list, representing the first association rule found 
(there could be more if multiple rules were discovered for the itemset).

result[2][0][0]: This accesses the items_base of the first ordered statistic, which corresponds to the antecedent itemset 
(the left-hand side of the association rule). Note that items_base is also a frozenset representing the antecedent itemset.

result[2][0][1]: This part of the indexing is used to access the consequent (right-hand side) itemset of the association rule.
Let's break it down:

[0]: Finally, after accessing the frozenset representing the antecedent or consequent itemset, 
[0] is used to retrieve the first (and in this case, only) item in that itemset. 
Since both the antecedent and consequent itemsets are single-item sets in this case, [0] allows us to access that
single item directly."""

def inspect(results):
    product1         = [tuple(result[2][0][0])[0] for result in results]
    product2        = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(product1, product2, supports, confidences, lifts))
DataFrame_intelligence = pd.DataFrame(inspect(results), columns = ['product1', 'product1', 'Support', 'Confidence', 'Lift'])
```
